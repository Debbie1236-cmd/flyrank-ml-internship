## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Decision: Which pages on a client's site should be reviewed first for a content refresh, given limited team capacity?
Unit of analysis: one page (one row = one content item, content_id).
Who acts on it: a content/marketing team member, who decides whether to update, expand, or leave a flagged page.
Action taken: rewriting outdated information, updating stats, improving structure, or otherwise refreshing the page.
Cost of a wrong call: if the system flags a page that's actually fine, the team wastes limited review time (a false alarm). If it misses a page that's quietly losing real visitors, the business loses traffic and potential customers without anyone noticing until it's too late (a missed problem). Both mistakes cost something different — wasted effort vs. lost opportunity — so the ranking needs to balance both.

Using the starter dataset, I'm checking two numbers: what share of pages are currently flagged as declining, and how click-through rate changes by search position — both suggest there's real, non-trivial variation worth ranking pages on.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Number 1: how many pages are currently declining
declining_rate = (df["trend_direction"] == "down").mean()
print(f"Share of pages currently declining: {declining_rate:.1%}")

# Number 2: CTR gap between top and bottom position tiers
visible = df[df["impressions_90d"] >= 100]
ctr_by_tier = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print("\nAverage CTR by position tier:")
print(ctr_by_tier.round(3))
```
